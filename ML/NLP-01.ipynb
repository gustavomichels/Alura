```python
# Imports

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from nltk import tokenize
import seaborn as sns
import string
from nltk import RSLPStemmer
```

```python
# Dataset

primary_data = pd.read_csv('Dados/NLP/imdb-reviews-pt-br.csv')
```

```python
# Checking dataset

primary_data
```

```python
# Replacing sentiments to 0 and 1 (negative & positive)

primary_data['classification'] = primary_data.sentiment.replace(['neg','pos'], [0,1])
primary_data
```

```python
# Testing count vectorizer

text = ['Watched a good movie', 'Watched a bad Movie']

vectorize = CountVectorizer(lowercase=False)
bag_of_words = vectorize.fit_transform(text)
bag_of_words
```

```python
# Getting text dictionary

vectorize.get_feature_names_out()
```

```python
# Creating text matrix

matrix_sparse  = pd.DataFrame.sparse.from_spmatrix(bag_of_words, columns=vectorize.get_feature_names_out())
matrix_sparse
```

```python
# Applying on our dataset

vectorize = CountVectorizer(lowercase=False, max_features=50)
bag_of_words = vectorize.fit_transform(primary_data.text_en)


train, test, train_sentiment, test_sentiment = train_test_split(bag_of_words, primary_data.classification, random_state=42)

logistic_regression = LogisticRegression()
logistic_regression.fit(train, train_sentiment)
accuracy = logistic_regression.score(test, test_sentiment)

print(accuracy)
```

```python
# Creating function to use on next tests

def classify_text(dataset, column, classification):
    vectorize = CountVectorizer(lowercase=False, max_features=50)
    bag_of_words = vectorize.fit_transform(dataset[column])


    train, test, train_sentiment, test_sentiment = train_test_split(bag_of_words, dataset[classification], random_state=42)

    logistic_regression = LogisticRegression()
    logistic_regression.fit(train, train_sentiment)
    return logistic_regression.score(test, test_sentiment)

print(classify_text(primary_data, 'text_en', 'classification'))
```

```python
# Creating variable to storage all words

words = ''.join([text for text in primary_data.text_en])

# Creating word cloud

word_cloud = WordCloud(width=800, height=500, max_font_size=110, collocations=False).generate(words)

# Plotting

plt.figure(figsize= (10,7))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('Off')
plt.show()
```

```python
# Separating in good ratings and bad ratings

def bad_ratings_cloud(text, column):
    
    text_bad = text.query('sentiment == "neg"')
    

    words = ''.join([text for text in text_bad[column]])
    
    # Creating word cloud
    
    word_cloud = WordCloud(width=800, height=500, max_font_size=110, collocations=False).generate(words)
    
    # Plotting
    
    plt.figure(figsize= (10,7))
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis('Off')
    plt.show()

def good_ratings_cloud(text, column):
    
    text_good = text.query('sentiment == "pos"')
    

    words = ''.join([text for text in text_good[column]])
    
    # Creating word cloud
    
    word_cloud = WordCloud(width=800, height=500, max_font_size=110, collocations=False).generate(words)
    
    # Plotting
    
    plt.figure(figsize= (10,7))
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.axis('Off')
    plt.show()
```

```python
# Bad ratings wordcloud

bad_ratings_cloud(primary_data, 'text_en')
```

```python
# Good ratings wordcloud

good_ratings_cloud(primary_data, 'text_en')
```

```python
# Tokenize data

tokenize_space = tokenize.WhitespaceTokenizer()
tokenize_words = tokenize_space.tokenize(words)
frequency = nltk.FreqDist(tokenize_words)


# Creating dataframe for frequency

df_frequency = pd.DataFrame({'Word': list(frequency.keys()), 'Frequency': list(frequency.values())})
df_frequency
```

```python
# Creating function to plot pareto

def pareto_plot(df, column, quantity):

    words = ''.join([text for text in df[column]])
    tokenize_words = tokenize_space.tokenize(words)
    frequency = nltk.FreqDist(tokenize_words)
    df_frequency = pd.DataFrame({'Word': list(frequency.keys()), 'Frequency': list(frequency.values())})
    df_frequency = df_frequency.nlargest(columns='Frequency', n = quantity)
    plt.figure(figsize=(12,8))
    ax = sns.barplot(data=df_frequency, x='Word', y='Frequency', color='gray')
    ax.set(ylabel='Count')
    plt.show()
```

```python
# Plotting our data

pareto_plot(primary_data, 'text_en', 10)
```

```python
# Irrelevant words

# nltk.download('stopwords')

irrelevant_words = nltk.corpus.stopwords.words('english')
print(irrelevant_words)
```

```python
# Removing stopwords from our dataset

processed_phrases = list()

for rating in primary_data.text_en:
    new_phrase = list()
    text_words = tokenize_space.tokenize(rating)
    for word in text_words:
        if word not in irrelevant_words:
            new_phrase.append(word)
    processed_phrases.append(' '.join(new_phrase))

primary_data['treatment_01'] = processed_phrases
primary_data.head()
```

```python
# Classify new column

classify_text(primary_data, 'treatment_01', 'classification')
```

```python
# Plotting

pareto_plot(primary_data, 'treatment_01', 10)
```

```python
# Removing punctuation from the dataset

# First lets grab all the punctuation on a string

punctuation = list(string.punctuation)

punctuation_stop_words = punctuation + irrelevant_words

processed_phrases = list()

token_ponct = tokenize.WordPunctTokenizer()

# Now lets remove the punctuations form the treatment 01

for rating in primary_data['treatment_01']:
    new_phrase = list()
    text_words = token_ponct.tokenize(rating)
    for word in text_words:
        if word not in punctuation_stop_words:
            new_phrase.append(word)

    processed_phrases.append(' '.join(new_phrase))

primary_data['treatment_02'] = processed_phrases
primary_data
```

```python
# Checking accuracy

accuracy_treatment_02 = classify_text(primary_data, 'treatment_02', 'classification')
print(accuracy_treatment_02)
```

```python
# Setting all words to be lowercase

processed_phrases = list()

for rating in primary_data['treatment_02']:
    new_phrase = list()
    rating = rating.lower()
    text_words = token_ponct.tokenize(rating)
    for word in text_words:
        if word not in punctuation_stop_words:
            new_phrase.append(word)

    processed_phrases.append(' '.join(new_phrase))

primary_data['treatment_03'] = processed_phrases
primary_data
```

```python
# Checking accuracy

accuracy_treatment_03 = classify_text(primary_data, 'treatment_03', 'classification')
print(accuracy_treatment_03)
```

```python
# Checking pareto

pareto_plot(primary_data, 'treatment_03', 10)
```

```python
# nltk.download('rslp')

# Applying steemer to reduce to a better radial

stemmer = nltk.RSLPStemmer()

processed_phrases = list()

for rating in primary_data['treatment_03']:
    new_phrase = list()
    text_words = token_ponct.tokenize(rating)
    for word in text_words:
        new_phrase.append(stemmer.stem(word))

    processed_phrases.append(' '.join(new_phrase))

primary_data['treatment_04'] = processed_phrases
primary_data
```

```python
# Checking accuracy

accuracy_treatment_04 = classify_text(primary_data, 'treatment_04', 'classification')
print(accuracy_treatment_04)
```

```python
# Checking pareto

pareto_plot(primary_data, 'treatment_04', 10)
```

```python
# Applying Tfidf on classification without treatment to measure the weights of each word

tfidf = TfidfVectorizer(lowercase=False, max_features=50)

tfidf_brute = tfidf.fit_transform(primary_data['text_en'])
train, test, train_class, test_class = train_test_split(tfidf_brute, primary_data['classification'], random_state=42)

logistic_regression.fit(train, train_class)
accuracy_tfidf_brute = logistic_regression.score(test, test_class)

print(accuracy_tfidf_brute)
```

```python
# Applying to treated data

tfidf_treatment = tfidf.fit_transform(primary_data['treatment_04'])
train, test, train_class, test_class = train_test_split(tfidf_treatment, primary_data['classification'], random_state=42)

logistic_regression.fit(train, train_class)
accuracy_tfidf_treatment = logistic_regression.score(test, test_class)

print(accuracy_tfidf_treatment)
```

```python
# Using Tfidf with ngrams

tfidf = TfidfVectorizer(lowercase=False, ngram_range = (1,2))

tfidf_treatment = tfidf.fit_transform(primary_data['treatment_04'])
train, test, train_class, test_class = train_test_split(tfidf_treatment, primary_data['classification'], random_state=42)

logistic_regression.fit(train, train_class)
accuracy_tfidf_treatment_ngrams = logistic_regression.score(test, test_class)

print(accuracy_tfidf_treatment_ngrams)
```

```python
# Understanding what are the heaviest words in our positive dataset

weights = pd.DataFrame(
    logistic_regression.coef_[0].T,
    index = tfidf.get_feature_names_out()
)
weights.nlargest(10, 0)
```

```python
# Understanding what are the heaviest words in our negative dataset

weights = pd.DataFrame(
    logistic_regression.coef_[0].T,
    index = tfidf.get_feature_names_out()
)
weights.nsmallest(10, 0)
```
